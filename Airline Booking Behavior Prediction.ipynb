```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
# --- Prepare Data for Modeling ---
# Separate features and target variable
X = df.drop('booking_complete', axis=1)
y = df['booking_complete']
```

```python
# Identify categorical and numerical columns
categorical_cols = [col for col in X.columns if X[col].dtype == 'object']
numerical_cols = [col for col in X.columns if X[col].dtype in ['int64', 'float64']]
```

```python
numerical_transformer = StandardScaler()
```

```python
# Preprocessing for categorical columns
categorical_transformer = OneHotEncoder(handle_unknown='ignore')  # Let it handle categories
```

```python
# Create a column transformer to apply transformations to the correct columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
```

```python
# Create the pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])
```

```python
# Fit and transform the data
X_processed = pipeline.fit_transform(X)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, roc_curve
```

```python
# Model: Logistic Regression
model = LogisticRegression(solver='liblinear', class_weight='balanced', random_state=42)
model.fit(X_train, y_train)
y_pred_lr = model.predict(X_test)
y_prob_lr = model.predict_proba(X_test)[:, 1]
```

```python
# Model: Random Forest
rf_model = RandomForestClassifier(class_weight='balanced', random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
y_prob_rf = rf_model.predict_proba(X_test)[:, 1]
```

```python
def evaluate_model(y_true, y_pred, y_prob, model_name='Model'):
    print(f"--- {model_name} ---")
    print(classification_report(y_true, y_pred))
    print(f"AUC-ROC: {roc_auc_score(y_true, y_prob):.4f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_true, y_pred))

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    plt.figure(figsize=(6, 4))
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc_score(y_true, y_prob):.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.show()


evaluate_model(y_test, y_pred_lr, y_prob_lr, 'Logistic Regression')
evaluate_model(y_test, y_pred_rf, y_prob_rf, 'Random Forest')
```

```python
# Feature Importance (for Random Forest)
feature_importance = rf_model.feature_importances_
feature_names = pipeline.named_steps['preprocessor'].get_feature_names_out(X.columns)

importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importance})
importance_df = importance_df.sort_values(by='Importance', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df)
plt.title('Top 10 Feature Importance (Random Forest)')
plt.show()
```
